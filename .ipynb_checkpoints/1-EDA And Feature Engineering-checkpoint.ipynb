```python
df_country=pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
##To check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## Pie Chart- Top 3 countries that uses zomato
import matplotlib
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

Observation:Zomato maximum records or transaction are from India 
After that USA and then United Kingdoms

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 4.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 2.5 to 3.4----> average
5. when Rating is between 1.8 to 2.4----> Poor

```python
ratings.head()
```

```python
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','lightgreen','darkgreen'])
plt.show()
```

Observation:
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,hue='Rating color',legend=False,palette=['blue','red','orange','yellow','lightgreen','darkgreen'])
plt.show()
```
